```python
import re

intents = {
    "greeting": {
        "keywords": ["hi", "hello", "hey"],
        "response": "Hello! How can I help you today?"
    },
    "hours": {
        "keywords": ["hours", "open", "timing"],
        "response": "We are open Monday to Friday from 9 AM to 5 PM."
    },
    "location": {
        "keywords": ["location", "address", "where"],
        "response": "We are located in Dallas, Texas."
    },
    "services": {
        "keywords": ["services", "offer", "do"],
        "response": "We provide support, guidance, and basic information services."
    },
    "farewell": {
        "keywords": ["bye", "goodbye", "exit"],
        "response": "Goodbye! Have a nice day."
    }
}

def detect_intent(user_input):
    user_input = user_input.lower()
    for intent, data in intents.items():
        for word in data["keywords"]:
            if word in user_input:
                return intent, data["response"]
    return "unknown", "Sorry, I did not understand that."

print("Simple Chatbot Started. Type 'bye' to stop.")

while True:
    user_text = input("You: ")
    intent, response = detect_intent(user_text)
    print("Detected Intent:", intent)
    print("Bot:", response)

    if intent == "farewell":
        break
```
